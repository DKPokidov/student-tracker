# Пример работы с модулем учёта успеваемости

## 1. Импорт и создание курса

```python
from tracker import Course
from utils import read_students_from_file

# Добавляем студентов из файла
PATH = 'data/students_list.txt'
students = read_students_from_file(PATH)

# Создаём курс
course = Course("Python для начинающих", students=students)

course.get_all_students()
```

```python
# добавляет студента
course.add_student('Пупкин П. П.')
course.get_all_students()
```

```python
# удаляет студента
course.remove_student('Пупкин П. П.')
course.get_all_students()
```

## 2. Добавление тем

```python
# Добавляем темы
course.add_topic("Введение", "lecture", 0)
course.add_topic("Циклы", "practice", 15)
course.add_topic("Функции", "practice", 20)

print("Темы:", course.get_all_topics())
```

## 3. Выставление оценок и посещаемости

```python
# Оценки
course.set_grade("Александрова Д. М.", "Циклы", 12)
course.set_grade("Бокова П. И.", "Циклы", 14)
course.set_grade("Зайкова В. Д.", "Циклы", 8)

# Посещаемость
course.mark_attendance("Александрова Д. М.", "Циклы", True)
course.mark_attendance("Бокова П. И.", "Циклы", True)
course.mark_attendance("Зайкова В. Д.", "Циклы", False)

print("Оценки:", course.grades)
print("Посещаемость:", course.attendance)
```

## 4. Сохранение и загрузка данных

```python
# Сохраняем
course.save_to_file("data/example_course.json")
print("✅ Данные сохранены")

# Загружаем в новый курс
new_course = Course("Загруженный курс")
new_course.load_from_file("data/example_course.json")

print("✅ Данные загружены")
print("Студенты:", new_course.students)
print("Темы:", new_course.topics)
print("Оценки:", new_course.grades)
```

## 5. Отчёты

```python
# Средний балл студента
print(f"Средний балл Иванова: {course.get_student_average('Александрова Д. М.')}")

# Средний балл группы
print(f"Средний балл группы: {course.get_group_average()}")

# Должники по теме
debtors = course.get_debtors("Циклы", 10)
print(f"Должники по теме 'Циклы': {debtors}")

# Посещаемость студента
rate = course.get_attendance_rate("Александрова Д. М.")
print(f"Посещаемость Александрова Д. М.: {rate}%")
```

## 6. Сводная таблица оценок

```python
def show_grades_table(course):
    """Выводит сводную таблицу оценок"""
    if not course.grades:
        print("📭 Оценок пока нет.")
        return
    
    print("\n📊 Сводная таблица оценок:")
    all_students = sorted(course.students)
    all_topics = [topic['topic_name'] for topic in course.topics]
    
    # Заголовок
    header = "Студент".ljust(25)
    for topic in all_topics:
        header += topic.ljust(15)
    print(header)
    print("-" * (25 + 15 * len(all_topics)))
    
    # Данные
    for student in all_students:
        row = student.ljust(25)
        grades = course.get_all_grades(student)
        for topic in all_topics:
            score = grades.get(topic, "-")
            row += str(score).ljust(15)
        print(row)

# Использование
show_grades_table(course)
```
